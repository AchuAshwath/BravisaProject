```python
# Given list of strings
import json

for column in industrymap.columns:
    if column in ['Industry', 'SubSector', 'Sector', 'SubIndustry']:
        categories = industrymap[column].unique()
        if column == 'SubIndustry':
            # if ['Industrial Gases', 'Cement & Products', 'Construction & Engineering', 'Tires & Rubber' ,'Health Care Supplies'] in categories:
            #     #print these categories
            #     print("found")
            # print(categories)
            pass
        # Initialize an empty dictionary
        category_to_caps = {}

        # Iterate over the list
        for category in categories:
            # Extract capital letters from the string
            caps = ''.join(filter(str.isupper, category))
            # If caps consist of just a single letter, get the first three letters of the string
            if len(caps) == 1:
                caps = category[:3]
                caps = caps.upper()
            if caps in category_to_caps:
                # print(caps)
                caps = category[:4].upper()
                # print(caps)
                category_to_caps[caps] = category   
            # Map the capital letters to the original string in the dictionary
            category_to_caps[caps] = category   
        
        reversed_dict = {}

        # Iterate over the items in the original dictionary
        for key, value in category_to_caps.items():
            # Swap the key and value and add to the reversed dictionary
            reversed_dict[value] = key

        if len(industrymap[column].unique())==len(set(category_to_caps.keys())):
            # print(json.dumps(reversed_dict, indent=4))   
            for index, row in industrymap.iterrows():
                if row[column] in reversed_dict.keys():
                    # create a new column called column+index
                    industrymap.at[index, column+'IndexName'] = reversed_dict[row[column]]
                    # print(row[column], reversed_dict[row[column]])
        else:
            print("Error in mapping for column : ", column)
            print(len(industrymap[column].unique()), len(set(category_to_caps.keys())))
```

```python
# duplicated of industrymap
industrymap_dup = industrymap[industrymap.duplicated(subset=['Industry', 'SubSector', 'Sector', 'SubIndustry', 'IndustryCode'], keep=False)]

# drop industrymap_dup
industrymap.drop_duplicates(subset=['Industry', 'SubSector', 'Sector', 'SubIndustry', 'IndustryCode'], keep="first", inplace=True)
```
